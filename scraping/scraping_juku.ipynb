```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
```

```python
f = urlopen('http://www.yozemi.ac.jp/moshi/schedule/')
soup = BeautifulSoup(f, 'html.parser')
soup.title
```

```python
soup.html
```

```python
lnkObj = soup.findAll("p", {"class":"mod-link"})
```

```python
for ls in lnkObj:
    for l in ls.findAll("a"):
        print("test : {} usl : {}".format(l.get_text(), l.attrs['href']))
```
